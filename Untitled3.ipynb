```python
wcss =[]

for i in range (1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss,linewidth=2)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```
